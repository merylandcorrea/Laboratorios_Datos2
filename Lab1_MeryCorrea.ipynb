```python
from typing import Any, List, Optional, Tuple
import matplotlib.pyplot as plt
import networkx as nx
```

```python
class Stack:

    def __init__(self) -> None:
        self.stack: List[Any] = []

    def add(self, elem: Any) -> None:
        self.stack.append(elem)

    def remove(self) -> Any:
        return self.stack.pop()

    def is_empty(self) -> bool:
        return len(self.stack) == 0
```

```python
class Queue:

    def __init__(self) -> None:
        self.queue: List[Any] = []

    def add(self, elem: Any) -> None:
        self.queue.append(elem)

    def remove(self) -> Any:
        return self.queue.pop(0)

    def is_empty(self) -> bool:
        return len(self.queue) == 0
```

```python
class Node:

   def __init__(self, data: Any) -> None:
        self.data = data
        self.left: Optional["Node"] = None
        self.right: Optional["Node"] = None
        self.height: int = 1
```

```python
from ast import Return
class AVLTree:

    def __init__(self, root: "Node" = None) -> None:
        self.root = root

    def insert(self, elem: Any) -> bool:
        to_insert = Node(elem)
        if self.root is None:
            self.root = to_insert
            return True
        else:
            p, pad = self.search(elem)
            if p is None:
                if elem < pad.data:
                    pad.left = to_insert
                else:
                    pad.right = to_insert
                self.alturaNodo() #Da la altura del nodo

                return True
            return False

    def alturaNodo(self) -> None:
        self.root.height=1
        p, q = self.root, Queue()

        q.add(p)
        while not q.is_empty():
            p = q.remove()
            if p.left is not None:
                p.left.height = p.height + 1
                q.add(p.left)
            if p.right is not None:
                p.right.height = p.height + 1
                q.add(p.right)

    def rotaciones(self,pad,balance,FE_Pad):
        if pad.left!= None:
          if balance == -2 and FE_Pad == -1:
            print('RSD')
            self.simple_right_rotation(pad)
          if balance == -2 and FE_Pad == 1:
            print('RDID')
            self.double_left_right_rotation(pad)

        if pad.right != None:
         if balance == 2 and FE_Pad == 1:
            print('RSI')
            self.simple_left_rotation(pad)
         if balance == 2 and FE_Pad == -1:
            print('RSDI')
            self.double_right_left_rotation(pad)



    def simple_left_rotation(self, node: "Node"):
      temp = self.search_father(node.data)
      aux = node.right
      node.right = aux.left
      aux.left = node
      if temp == None:
        self.root = aux
      elif node.data < temp.data and temp !=None:
         temp.left = aux
      else:
         temp.right = aux
      return aux

    def simple_right_rotation(self, node: "Node"):
      temp = self.search_father(node.data)
      aux = node.left
      node.left = aux.right
      aux.right = node
      if temp == None:
        self.root = aux
      elif node.data < temp.data and temp !=None:
         temp.left = aux
      else:
         temp.right = aux
      return aux

    def double_right_left_rotation(self, node: "Node"):
      self.simple_right_rotation(node.right)
      return self.simple_left_rotation(node)

    def double_left_right_rotation(self, node: "Node"):
      self.simple_left_rotation(node.left)
      return self.simple_right_rotation(node)

     # se hace un recorrido por nivel verificando el sub arbol derecho e izquierdo
     # con la variable contador guardo la difirencia del nivel del ultimo nodo con respecto a la raiz que es el nodo
     # del cual se desea saber su factor de balanceo
     #por ultimo este devuelve el valor de la resta del sub arbol derecho menos el izquierdo
    def factor_balanceo(self , node: Optional[Node]) -> None:
        p, s = node, Queue()
        contador_izq = 0
        contador_der = 0
        s.add(p)
        while not s.is_empty():
            p = s.remove()
            # print(p.data)
            if p.left is not None:
                if p.left.data < node.data:
                  contador_izq = p.left.height - node.height
                elif p.left.data > node.data:
                  contador_der = p.left.height - node.height
                s.add(p.left)

            if p.right is not None:
                if p.right.data > node.data:
                  contador_der = p.right.height - node.height
                elif p.right.data < node.data:
                  contador_izq = p.right.height - node.height
                s.add(p.right)

        return contador_der - contador_izq




    def delete(self, elem: Any, mode: bool = True) -> bool:
        p, pad = self.search(elem)

        if p is not None:
            if p.left is None and p.right is None:
                if p == pad.left:
                    pad.left = None
                else:
                    pad.right = None
                del p
            elif p.left is not None and p.right is None:
                if p == pad.left:
                    pad.left = p.left
                else:
                    pad.right = p.left
                del p
            elif p.left is None and p.right is not None:
                if p == pad.left:
                    pad.left = p.right
                else:
                    pad.right = p.right
                del p
            else:
                if mode:
                    pred, pad_pred = self.__pred(p)
                    p.data = pred.data
                    if pred.left is not None:
                        if pad_pred == p:
                            pad_pred.left = pred.left
                        else:
                            pad_pred.right = pred.left
                    else:
                        if pad_pred == p:
                            pad_pred.left = None
                        else:
                            pad_pred.right = None
                    del pred
                else:
                    sus, pad_sus = self.__sus(p)
                    p.data = sus.data
                    if sus.right is not None:
                        if pad_sus == p:
                            pad_sus.right = sus.right
                        else:
                            pad_sus.left = sus.right
                    else:
                        if pad_sus == p:
                            pad_sus.right = None
                        else:
                            pad_sus.left = None
                    del sus
            return True
        return False

    def __pred(self, node: "Node") -> Tuple["Node", "Node"]:
        p, pad = node.left, node
        while p.right is not None:
            p, pad = p.right, p
        return p, pad

    def __sus(self, node: "Node") -> Tuple["Node", "Node"]:
        p, pad = node.right, node
        while p.left is not None:
            p, pad = p.left, p
        return p, pad



    def search_nodo(self, elem: Any) -> bool:
        p, pad = self.root, None
        while p is not None:
            if elem == p.data:
                return True
            elif elem < p.data:
                pad = p
                p = p.left
            else:
                pad = p
                p = p.right

        return False

    def BuscarGeneral(self, elem: int) -> Optional[Node]:
        p = self.root
        while p is not None:
            if elem == p.data:
                return p
            elif elem < p.data:
                p = p.left
            else:
                p = p.right
        return None

    def search(self, elem: Any) -> Tuple[Optional["Node"], Optional["Node"]]:
        p, pad = self.root, None
        while p is not None:
            if elem == p.data:
                return p, pad
            elif elem < p.data:
                pad = p
                p = p.left
            else:
                pad = p
                p = p.right

        return p, pad

    def search_father(self, data_s: Any) -> None:
        p, pad = self.root, None
        s, flag = Stack(), False
        while (p is not None or not s.is_empty()) and not flag:
            if p is not None:
                if p.data == data_s:
                    if pad is not None:
                        return pad
                    flag = True
                else:
                    s.add(p)
                    pad = p
                    p = p.left
            else:
                p = s.remove()
                pad = p
                p = p.right

        if not flag or pad is None:
            return None


    def search_abuelo(self, node_data: Any) -> Optional[Node]:
        node, parent = self.search(node_data)
        if parent is None:
            return None  # Si el nodo no tiene padre, no tiene abuelo
        grandparent_data = parent.data
        _, grandparent = self.search(grandparent_data)  # Busca el nodo con el dato del abuelo
        return grandparent  # Retorna el nodo del abuelo del nodo dado


    def search_tio(self, node1, node2, node3, elem):
        if node1 is not None:
          if elem == node1.data:
              return node3
          elif elem < node1.data:
              if node2 is not None:
                  if node1 == node2.left:
                      return self.search_tio(node1.left, node1, node2.right, elem)
                  else:
                      return self.search_tio(node1.left, node1, node2.left, elem)
              else:
                  return self.search_tio(node1.left, node1, None, elem)
          else:
              if node2 is not None:
                  if node1 == node2.left:
                      return self.search_tio(node1.right, node1, node2.right, elem)
                  else:
                      return self.search_tio(node1.right, node1, node2.left, elem)
              else:
                  return self.search_tio(node1.right, node1, None, elem)
        else:
          return None



    def levels_nr(self) -> None:
        p, q = self.root, Queue()
        q.add(p)
        while not q.is_empty():
            p = q.remove()
            print(p.data, end = ' ')
            if p.left is not None:
                q.add(p.left)
            if p.right is not None:
                q.add(p.right)



    def build_graph(self, graph, node, pos, x_pos, y_pos, y_step):
        if node:
            graph.add_node(node.data)
            pos[node.data] = (x_pos, y_pos)
            if node.left:
                graph.add_edge(node.data, node.left.data)
                self.build_graph(graph, node.left, pos, x_pos - y_step, y_pos - 1, y_step / 2)
            if node.right:
                graph.add_edge(node.data, node.right.data)
                self.build_graph(graph, node.right, pos, x_pos + y_step, y_pos - 1, y_step / 2)


    def draw(self):
       G = nx.DiGraph()
       pos = {}
       self.build_graph(G, self.root, pos, 0, 0, 1)

       plt.figure(figsize=(10, 8))
       nx.draw(G, pos, with_labels=True, node_size=200, node_color="skyblue", font_size=6, font_weight='bold', arrows=True, arrowstyle='->')
       plt.title("Árbol AVL")
       plt.show()
```
